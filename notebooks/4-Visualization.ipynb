```python
#import libraries
import os
import pandas as pd
import random
import numpy as np
import math
import scipy.stats as st
from statistics import stdev
from scipy.stats import ttest_ind_from_stats
from datetime import date, timedelta
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
get_ipython().run_line_magic('matplotlib', 'inline')
sns.set(color_codes=True)

```

```python
 #Import a CSV File into Python using Pandas
#define working directory
os.chdir("../datasets/stock_data/")
#read csv file 
AB_df = pd.read_csv("AB.csv")
```

```python
current_date = date.today().isoformat()   
days_before = (date.today()-timedelta(days=30)).isoformat()
fig = plt.figure(figsize=(30,10))
ax = fig.add_axes([0,0,1,1])
volum = AB_df['Volume']
days= AB_df['Date']
ax.bar(days,volum)
plt.xlabel("date")
plt.ylabel("Volum")
plt.title("AB Stock Volume from "+ days_before +" to " + current_date )
plt.show()
```

```python

AB_df[['Close']].plot(label='AB',figsize=(16,8),title='ClosePrice')
plt.legend()
plt.xlabel("date")
plt.ylabel("$ price")
plt.title("AB Stock Price from "+ days_before +" to " + current_date )
 
plt.show()
```
